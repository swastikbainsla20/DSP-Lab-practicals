```python
from nltk.corpus import brown
from sklearn.model_selection import train_test_split
from collections import defaultdict, Counter
import numpy as np

# Load the Brown corpus with the universal tagset

brown_tagged_sents = brown.tagged_sents(tagset='universal')
train_sents, test_sents = train_test_split(brown_tagged_sents, test_size=0.2)

# Prepare the data structures for HMM probabilities
tag_word_counts = defaultdict(Counter)
tag_counts = Counter()
tag_bigram_counts = defaultdict(Counter)

# 1. Data Preparation
print("Training data preparation...")

for sent in train_sents:
    previous_tag = "<START>"
    for word, tag in sent:
        word = word.lower()  # Convert to lowercase for consistency
        tag_word_counts[tag][word] += 1
        tag_counts[tag] += 1
        tag_bigram_counts[previous_tag][tag] += 1
        previous_tag = tag
    tag_bigram_counts[previous_tag]["<END>"] += 1

print(f"Number of unique tags: {len(tag_counts)}")
print(f"Tag counts: {tag_counts.most_common(5)}")
print(f"Word counts for tag 'NOUN': {tag_word_counts['NOUN'].most_common(5)}")
```

```python
# 2. Implementing HMM with Smoothing
def emission_prob(word, tag, k=0.1):  # Add-one smoothing
    # Emission probability with smoothing
    return (tag_word_counts[tag][word] + k) / (tag_counts[tag] + k * len(tag_word_counts[tag]))

def transition_prob(prev_tag, current_tag, k=0.1):  # Add-one smoothing
    # Transition probability with smoothing
    return (tag_bigram_counts[prev_tag][current_tag] + k) / (sum(tag_bigram_counts[prev_tag].values()) + k * len(tag_bigram_counts[prev_tag]))

# Test emission and transition probabilities for sample words
print(f"Emission probability for 'dog' given tag 'NOUN': {emission_prob('dog', 'NOUN')}")
print(f"Transition probability from 'DET' to 'NOUN': {transition_prob('DET', 'NOUN')}")
```

```python
# 3. Viterbi Algorithm
def viterbi(sentence, tagset):
    viterbi_matrix = [{}]
    backpointer = [{}]

    # Initialize base case
    print(f"\nProcessing sentence: {' '.join(sentence)}")
    for tag in tagset:
        viterbi_matrix[0][tag] = transition_prob("<START>", tag) * emission_prob(sentence[0].lower(), tag)
        backpointer[0][tag] = None
        print(f"Initial probabilities for tag '{tag}' with word '{sentence[0]}': {viterbi_matrix[0][tag]}")

    # Run Viterbi for the rest of the words
    for i in range(1, len(sentence)):
        viterbi_matrix.append({})
        backpointer.append({})

        for tag in tagset:
            max_prob, best_prev_tag = max(
                (viterbi_matrix[i-1][prev_tag] * transition_prob(prev_tag, tag) * emission_prob(sentence[i].lower(), tag), prev_tag)
                for prev_tag in tagset
            )
            viterbi_matrix[i][tag] = max_prob
            backpointer[i][tag] = best_prev_tag
            print(f"Probability for tag '{tag}' with word '{sentence[i]}': {viterbi_matrix[i][tag]} from previous tag '{best_prev_tag}'")

    # Backtrack to find the best sequence
    best_tags = []
    max_prob, best_final_tag = max((viterbi_matrix[-1][tag], tag) for tag in tagset)
    best_tags.append(best_final_tag)

    for i in range(len(sentence) - 1, 0, -1):
        best_tags.insert(0, backpointer[i][best_tags[0]])

    print(f"Best tag sequence for the sentence: {best_tags}")
    return best_tags
```

```python
# 4. Evaluation Function
def evaluate(test_sents, tagset):
    correct = total = 0
    for sent in test_sents:
        words = [word for word, tag in sent]
        true_tags = [tag for word, tag in sent]
        predicted_tags = viterbi(words, tagset)

        for true_tag, predicted_tag in zip(true_tags, predicted_tags):
            if true_tag == predicted_tag:
                correct += 1
            total += 1
    accuracy = correct / total
    print(f"\nPOS Tagger Accuracy: {accuracy * 100:.2f}%")
    return accuracy
```

```python
# Set of all tags
tagset = list(tag_counts.keys())

# Run evaluation on a small subset of test data for demonstration
print("Evaluating the model...")
accuracy = evaluate(test_sents[:50], tagset)  # Evaluate on a small set to test print output
```

```python
import nltk
from nltk import CFG

# Define a context-free grammar for English sentences
grammar = CFG.fromstring("""
    S -> NP VP
    NP -> Det N | Det Adj N | N
    VP -> V NP | V Adv NP | V PP | V Adv | Adv V NP
    PP -> P NP
    Det -> 'the' | 'a'
    N -> 'cat' | 'dog' | 'mat' | 'John' | 'apple' | 'dog'
    Adj -> 'quick' | 'lazy' | 'happy'
    V -> 'saw' | 'ate' | 'sat' | 'likes' | 'chased'
    P -> 'on' | 'under'
    Adv -> 'quickly' | 'silently'
""")

print(grammar)
```

```python
from nltk.parse import ChartParser

# Initialize the parser with the defined grammar
parser = ChartParser(grammar)

def check_sentence(sentence):
    print(f"\nChecking sentence: {' '.join(sentence)}")
    try:
        # Try parsing the sentence
        parse_trees = list(parser.parse(sentence))

        # If no parse trees are found, the sentence is invalid
        if len(parse_trees) == 0:
            print("Invalid sentence!")
        else:
            print("Valid sentence!")
            for tree in parse_trees:
                tree.pretty_print()  # Show the parse tree

    except ValueError:
        print("Invalid sentence!")
```

```python
sentences_part1 = [
    "the quick dog silently ate the apple",
    "the cat sat on the mat",
    "John likes the happy dog",
    "cat the mat on"
]
# Function to test a list of sentences
def test_sentences(sentences):
    for sentence in sentences:
        check_sentence(sentence.split())

# Run the test
test_sentences(sentences_part1)
```

```python
sentences_part2 = ["John the apple likes",
    "sat on mat the",
    "a dog likes John",
    "the lazy cat ate quickly",
    "quickly John the ate"]

# Function to test a list of sentences
def test_sentences(sentences):
    for sentence in sentences:
        check_sentence(sentence.split())

# Run the test
test_sentences(sentences_part2)
```
